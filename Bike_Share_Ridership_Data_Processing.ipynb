```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime as dt
```

```python
# Download all datasets from Toronto Bike Share Ridership to be merged
dataset_names = ['2019q1','2019q2','2019q3','2019q4',
                '2020-01','2020-02','2020-03','2020-04','2020-05','2020-06','2020-08','2020-07','2020-09','2020-10','2020-11','2020-12',
                'Bike share ridership 2021-01','Bike share ridership 2021-02','Bike share ridership 2021-03','Bike share ridership 2021-04','Bike share ridership 2021-05','Bike share ridership 2021-06','Bike share ridership 2021-07','Bike share ridership 2021-08','Bike share ridership 2021-09','Bike share ridership 2021-10','Bike share ridership 2021-11','Bike share ridership 2021-12',
                'Bike share ridership 2022-01','Bike share ridership 2022-02','Bike share ridership 2022-03','Bike share ridership 2022-04','Bike share ridership 2022-05','Bike share ridership 2022-06','Bike share ridership 2022-07','Bike share ridership 2022-08','Bike share ridership 2022-09','Bike share ridership 2022-10','Bike share ridership 2022-11','Bike share ridership 2022-12',
                'Bike share ridership 2023-01','Bike share ridership 2023-02','Bike share ridership 2023-03','Bike share ridership 2023-04','Bike share ridership 2023-05','Bike share ridership 2023-06','Bike share ridership 2023-07','Bike share ridership 2023-08','Bike share ridership 2023-09','Bike share ridership 2023-10','Bike share ridership 2023-11','Bike share ridership 2023-12']
```

```python
# Some files are encoded differently and cannot be opened with csv
dfs= []
common_columns = None
for dataset_name in dataset_names:
    file_path = f"/content/drive/MyDrive/SAS 2024 Safe Roads/{dataset_name}.csv"  # Replace with the actual file path
    print(dataset_name)
    encodings = ["utf-8", "latin-1", "utf-16"]
    for encoding in encodings:
        try:
            df = pd.read_csv(file_path, encoding=encoding)
            break
        except UnicodeDecodeError:
            pass
    if common_columns is None:
        # For the first dataset, store the column names
      common_columns = df.columns
    else:
        # Reorder the columns of the new DataFrame to match the existing DataFrame
      df = df.reindex(columns=common_columns)
    dfs.append(df)
```

```python
#Merge all dataframes
merged_df = pd.concat(dfs)
```

```python
merged_df.to_csv('/content/drive/MyDrive/SAS 2024 Safe Roads/mergeddataset.csv',index=False)
```
